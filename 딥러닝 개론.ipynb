## 경사하강법 구현

```python
import numpy as np
x=[2,4,6,8] # 문제지
y=[20,55,50,80] # 답안지
```

```python
x=np.array(x)
y=np.array(y)
```

```python
# 경사하강법 구현해보기
#learning rate를 최적화 해주는 함수 Adam**
a=0
b=0
lr=1 # learning rate

for n in range(3000):
    y_hat=a*x+b # 에측선 그리기
    error=y_hat-y

    a_diff=sum(2*error*x) # 미분된 식을 다 더하는
    b_diff=sum(2*error*1)

    a=a-a_diff
    b=b-b_diff

    print("{}회 학습//{},{}".format(n+1,a_diff,b_diff))
    # 결과값을 확인해보니, 값이 커지고 있음을 확인 -> a와 b 모두 잘못 계산되고 있음
```

```python
# learning rate를 조정
#learning rate를 최적화 해주는 함수 Adam**
a=0
b=0
lr=0.00# learning rate

for n in range(3000):
    y_hat=a*x+b # 에측선 그리기
    error=y_hat-y

    a_diff=sum(2*error*x) # 미분된 식을 다 더하는
    b_diff=sum(2*error*1)

    a=a-a_diff*lr
    b=b-b_diff*lr

    print("{}회 학습//{:.2f},{:.2f}".format(n+1,a_diff,b_diff))
    # 결과값을 확인해보니, 값이 커지고 있음을 확인 -> a와 b 모두 잘못 계산되고 있음
```

```python
x=np.arange(-3,3,0.01)
y=np.sin(x)
lr=0.00001# learning rate
a=0
b=0
c=0
d=0

for n in range(3000):
    y_hat=a*x**3+b*x**2+c*x+d # 에측선 그리기
    error=y_hat-y

    a_diff=sum(2*error*x**3) # 미분된 식을 다 더하는, a_diff a의 기울기
    b_diff=sum(2*error*x**2)
    c_diff=sum(2*error*x)
    d_diff=sum(2*error*1)

    #a,b,c,d의diff들이 0이 되는 a,b,c,d,값을 찾기
    a=a-a_diff*lr
    b=b-b_diff*lr
    c=c-c_diff*lr
    d=d-d_diff*lr

    print("{}회 학습//{:.2f},{:.2f},{:.2f},{:.2f}".format(n+1,a_diff,b_diff,c_diff,d_diff))
    # 결과값을 확인해보니, 값이 커지고 있음을 확인 -> a와 b 모두 잘못 계산되고 있음
```

```python
import pandas as pd
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
pima=pd.read_csv('pima_indians.csv')
bike=pd.read_csv('bike.csv')
```

```python
#weather 1-좋음
bike
```

```python
del bike['registered']
del bike['datetime']
```

```python
# 문제지
data=bike.iloc[:,:-1].to_numpy()
# 정답지
target = bike['count'].to_numpy()
```

```python
train_input,test_input,train_target,test_target=train_test_split(data,target,train_size=0.8
                                                                 )
```

```python
#표준화
ss=StandardScaler()
train_scaled=ss.fit_transform(train_input)
test_scaled=ss.transform(test_input)
```

```python
train_scaled.shape
```

```python
model=keras.Sequential() #인공신경망을 만들기 위한 흰 도화지
model.add(keras.layers.Dense(1,input_shape=(9,))) # 도화지에다가 10개의 input과
```

```python
model.summary() # 현재 컴퓨터가 학습해야하는 parameter은 10개
```

```python
# 모델에게 학습 방법을 알려주고
model.compile(loss='mse',optimizer = 'adam',metrics='mae')
# 오차 계산은 mse로 선택하고, learning rate(학습률)은 자동으로 최적화하고 우리에게는 mae값을 보여줘
```

```python
# 실제 데이터를 줘서 학습시키는
es=keras.callbacks.EarlyStopping(patience=5,restore_best_weights=True)
#patience: 과대적합을 몇번까지 허용할 것인지(원래는 20~30 정도), restore_best_weights는 과대적합이 시작된 부분에서 멈추고 확인하는 (5번까지 과대적합이 일어나도 모두 확인하는 것이 아닌, 처음값만 확인하는)
hist=model.fit(train_scaled,train_target,epochs=100,validation_data=(test_scaled,test_target),
               callbacks=[es])
# 학습하고 쪽지시험을 보는 것 -> 만약 공부할때 점수는 안 좋은데 쪽시 시험을 봤을 때 점수가 높다면 이것이 과대적합
# 과대학습을 방지하기 위해 과대적합이 일어날 지점에서 학습을 멈춰주는 early_stoping을 사용하면 된다.

# epoch 밑에 숫자는 컴퓨터가 눈이 1개가 아니라 32개여서 데이터가 1000개 라면 눈이 32개여서 한번에 30번만 보면 된다.
```

```python
plt.plot(hist.history['loss']) # 검증 데이터의 오차값 확인
plt.plot(hist.history['val_loss']) # 검증 데이터의 오차값 확인
plt.legend(['loss','val_loss'])
plt.show()
# 어딘가로 수렴해야하는데 아직 학습횟수가 부족함을 볼 수 있음, 그리고 그림만 봐서는 정확히 알 수 없음
```

### 쉬운 데이터에 복잡한 모델을 넣어보기  
- 간단한 데이터에 복잡한 모델을 적용하게 되면 과적합이 쉽게 일어남.

```python
df = pd.read_csv("pima_indians.csv")

data = df.iloc[:, :-1].to_numpy()
target = df['Class'].to_numpy()

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size = 0.2)

ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)

model = keras.Sequential()
model.add(keras.layers.Dense(32, activation = 'sigmoid', input_shape = (8,))) #32개 노드를 가진  은닉층
model.add(keras.layers.Dense(16, activation = 'sigmoid', input_shape = (8,))) #16개 노드를 가진 은닉층
model.add(keras.layers.Dense(2, activation = 'softmax', input_shape = (8,))) # 가장 마지막 출력층은 우리가 보기 좋게 합이 1이 되는 확률값으로 출력

model.compile(loss = "sparse_categorical_crossentropy",
              optimizer = 'adam', metrics = 'accuracy')

es = keras.callbacks.EarlyStopping(patience=20, restore_best_weights=True)

hist = model.fit(train_scaled, train_target, epochs = 1000,
                 validation_data = (test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(hist.history['loss']) # 검증 데이터의 오차값 확인
plt.plot(hist.history['val_loss']) # 검증 데이터의 오차값 확인
plt.legend(['loss','val_loss'])
plt.show()
```

```python
df = pd.read_csv("pima_indians.csv")

data = df.iloc[:, :-1].to_numpy()
target = df['Class'].to_numpy()

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size = 0.2)

ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)

model = keras.Sequential()
model.add(keras.layers.Dense(10, activation = 'relu', input_shape = (8,))) #32개 노드를 가진  은닉층
model.add(keras.layers.Dropout(0,4)) # 32개 은닉층 중 랜덤으로 40% 선을 끊어버리는(50%가 넘으면 오히려 성능이 안 좋아질 수도 있다. 대부분 10~40%)
model.add(keras.layers.Dense(2, activation = 'softmax', input_shape = (8,))) # 가장 마지막 출력층은 우리가 보기 좋게 합이 1이 되는 확률값으로 출력

model.compile(loss = "sparse_categorical_crossentropy",
              optimizer = 'adam', metrics = 'accuracy')

es = keras.callbacks.EarlyStopping(patience=20, restore_best_weights=True)

hist = model.fit(train_scaled, train_target, epochs = 1000,
                 validation_data = (test_scaled, test_target))
```

```python
plt.plot(hist.history['loss']) # 검증 데이터의 오차값 확인
plt.plot(hist.history['val_loss']) # 검증 데이터의 오차값 확인
plt.legend(['loss','val_loss'])
plt.show()
```

```python
model.evaluate(hist)
```
